Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework written in Python that allows developers to quickly build web applications. It is a lightweight and flexible framework that provides a simple interface for developers to handle requests and responses, manage sessions, and more.

Advantages of Flask Framework include:

Lightweight and flexible: Flask is a micro-framework, which means it has a small core and is highly customizable. Developers can add the required libraries as per their needs, making the application lightweight and fast.

Easy to use: Flask has a simple and easy-to-use interface, making it ideal for beginners who are learning web development.

Extensible: Flask can be extended by adding extensions or libraries as per the project requirements. Flask has a large community of developers who have built many extensions and libraries that can be used to enhance the functionality of the application.

Built-in Development Server: Flask has a built-in development server that allows developers to test and debug their applications easily.

Large Community: Flask has a large community of developers who have built many plugins, libraries, and extensions that can be used to add features to the application.

Supports various databases: Flask supports various databases like PostgreSQL, MySQL, SQLite, and more, making it easy for developers to work with different types of databases.

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, routing refers to mapping URLs to functions. App routing is the process of defining URL patterns for a Flask web application and associating each URL with a Python function that will be executed when the URL is requested.

We use app routes in Flask to handle different HTTP requests and provide the appropriate response. By defining app routes, we can create different URL endpoints for our application and provide a customized response for each endpoint. For example, we can define an app route to handle the homepage, another app route to handle the login page, and so on.

The app routes are defined using the @app.route() decorator in Flask. The decorator takes the URL pattern as a parameter and associates it with the function defined immediately below the decorator.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

Flask, the url_for() function is used for URL building. This function generates a URL for the specified endpoint by taking into account the URL rules defined in the Flask application.

The url_for() function takes the name of the function as the first argument and any number of keyword arguments that represent the variable parts of the URL

from flask import Flask, request, url_for

app = Flask(__name__)

@app.route('/user/<username>')
def show_user_profile(username):
    return 'User %s' % username

@app.route('/')
def index():
    user_url = url_for('show_user_profile', username='John')
    return 'User URL: %s' % user_url

if __name__ == '__main__':
    app.run(debug=True)
